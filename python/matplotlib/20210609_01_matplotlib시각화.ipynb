#### 실습데이터 불러오기

```python
tips = sns.load_dataset('tips')
```

```python
tips
```

```python
# linegraph
plt.figure(figsize = (10,5),facecolor = 'gray')
tips.groupby('sex')['tip'].mean().plot()
```

```python
# barchart
plt.figure(figsize = (10,5),facecolor = 'white')
plt.style.use('classic')
#plt.style.use('ggplot')
tips.groupby('sex')['tip'].mean().plot.bar() 
```

## subplot

```python
plt.style.use('seaborn-poster')
plt.suptitle('Main Title',fontsize=20)
plt.subplot(2,2,1)
tips.groupby('sex')['tip'].mean().plot.bar()
plt.title('첫번째',fontsize = 12)
plt.subplot(2,2,2)
plt.title('B',fontsize = 12)
plt.subplot(2,2,3)
plt.title('세번째제목',fontsize = 12)
plt.subplot(2,2,4)
plt.title('D',fontsize = 12)
```

### subplot's'

```python
fig,axes = plt.subplots(2,2)

axes[0, 0].plot(1, 1)
axes[0, 1].plot(2, 2)
axes[1, 0].plot(3, 3)
axes[1, 1].plot(4, 4)
```

```python
fig.set_figwidth(5) # 너비조절
fig.set_figheight(10) # 높이조절
fig
```

```python
# subplot4개를 총 2개의 axes 각 2개씩 할당
fig,(axes1,axes2) = plt.subplots(2,2) 
```

```python
axes1[0].plot([1,2,3]) # subplot1
axes1[1].plot([-1,-2,-3]) # subplot2
axes2[0].plot([1,2,3])# subplot3
axes2[1].plot([-1,-2,-3])# subplot4
fig
```

```python
# subplot4개를 총 4개의 axes 각 1개씩 할당
fig,((axes1,axes2),(axes3,axes4)) = plt.subplots(2,2)
```

```python
axes1.plot([1,2,3]) #subplot1
axes2.plot([6,4,5]) #subplot2
axes3.plot([11,22,33])#subplot3
axes4.plot([-1,-2,-3])#subplot4
fig
```

### 성별에 따른 팁의 차이 그래프
