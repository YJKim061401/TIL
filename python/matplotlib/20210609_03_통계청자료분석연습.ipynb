```python
# 이제 모든 column 확인가능
df_kosis.head()
```

```python
# df.melt(id_vars = '') id_vars: Column(s) to use as identifier variables.
# 데이터 재구조화. id_vars(e.g. 시군구별)가 column으로 해당년도, 출생인구수가 variable로 들어감
df = df_kosis.melt(id_vars = '시군구별') 
df.head(60)
```

```python
# 위에 데이터 프레임보면 중간 '시군구별'들어가 있음 (필요없는 데이터).
# 이를 제거하기 위해 [!=] 사용 
df = df[df['시군구별'] != '시군구별'].copy()
```

```python
# 시군구별 빠졌는지 df.unique()로 확인하기
df['시군구별'].unique()
```

#### 필요한 데이터들로 가공하기 (년/월, 성별)

```python
print(type(df)) # 원래는 dataframe
print(df['variable']) # dataframe에서 variable만 떼서 확인
print(type(df['variable'])) # series로 리턴
```

```python
# series는 split 사용 불가능
# str을 붙여 stringmethods로 변경 후 split 사용
print(type(df['variable'].str.split('.')))
df['variable'].str.split('.')
```

```python
# 이제 (.)으로 구분되어진 값들 이름붙여서 원래 dataframe에 붙이기
# df['이름'] = df['variable'].str.split('.',expand = True)[0]
# expand = True 일 때, dataframe으로 리턴 / expand = False(default setting), series로 리턴
df['연도'] = df['variable'].str.split('.',expand = True)[0]
df['월'] = df['variable'].str.split('.',expand = True)[1]
df['성별'] = df['variable'].str.split('.',expand = True)[2]
```

```python
df.head()
```

```python
df.tail()
```

```python
df.info()
```

```python
# '성별' 분류값 unique로 확인
# none(Total), 1 (M), 2(F) 존재
df['성별'].unique()
```

```python
# nunique() : NaN을 뺀 unique 값 리턴 
# 1과 2 총 2개니까 2로 리턴
df['성별'].nunique()
```

```python
# df.fillna : NaN 값을 '전체'로 이름값 변경
# NaN 일때는 replace말고 fillna('변경할 값') 사용
df['성별'] = df['성별'].fillna('전체')
```

```python
# df.replace('원래값','변경할 값')
# 1과 2 각 남자, 여자 값으로 이름 변경 
df['성별'] = df['성별'].replace('1','남자').replace('2','여자')
```

```python
df['성별'].unique()
```

### 1. 전국 성별 전체 가져오기
* 조건 : isin 사용? --> 비교하는 대상이 같을 경우만 가능
* & | 와 같은 비교연산자 이용하여 가져와야함

```python
df_all = df[(df['시군구별'] == '전국') & (df['성별'] == '전체')]
print(df_all['시군구별'].unique())
print(df_all['성별'].unique())
df_all.head()
```

### 2. 지역별 출생아수 시각화

```python
# 전국을 뺀 나머지 지역별로 나눈 데이터
df_local = df[df['시군구별'] != '전국'].copy()
df_local.head()
```

```python
df_local['시군구별'].unique()
```

```python
# 성별 구분 없이 전체만 가져오기
df_local_all = df_local[df_local['성별'] == '전체'].copy()
df_local_all.head()
```

```python
df_local_all['성별'].unique()
```

```python
plt.figure(figsize = (20,4))
sns.pointplot(data = df_local,x = '연도',y = '출생아수',hue = '성별')
plt.figure(figsize = (20,4))
sns.pointplot(data = df_local_all,x = '연도',y = '출생아수',ci = None)
plt.figure(figsize = (20,4))
sns.pointplot(data = df_local_all,x = '연도',y = '출생아수',hue = '시군구별')
plt.legend(loc = 'center right',bbox_to_anchor = (1.17,0.5), ncol = 1)
```

### 3. 지역별 (서울, 경기도, 세종특별자치시)에 따른 출생아수 시각화

```python
# 여기서는 비교하는 대상이 같으므로(시군구별) isin 을 이용 가능
df_local_2 = df_local_all[df_local_all['시군구별'].isin(['서울특별시','경기도','세종특별자치시'])]
df_local_2['시군구별'].unique()
```

```python
plt.figure(figsize = (20,4))
sns.pointplot(data = df_local_2,x = '연도',y = '출생아수',hue = '시군구별')
plt.legend(loc = 'center right',bbox_to_anchor = (1.10,0.5), ncol = 1)
```
