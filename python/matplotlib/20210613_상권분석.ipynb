```python
# columns명으로 dataframe다시 만들기
dataset = df[['상호명','지점명','상권업종중분류명','시도명','시군구명','행정동명','도로명','경도','위도']]
```

```python
# 상권업종중분류명이 커피점/카페 인 것만 빼오기
df_coffee = dataset[dataset['상권업종중분류명'] == '커피점/카페']
```

```python
# 상호명이 스타벅스인것만 빼오기
df_starbucks = df_coffee.loc[df_coffee['상호명'].str.contains('스타벅스|starbucks|STARBUCKS')]
```

```python
# 상호명이 이디야인것만 빼오기
df_ediya = df_coffee.loc[df_coffee['상호명'].str.contains('이디야|ediya|EDIYA')]
```

### columns ('브랜드명') 하나 생성해서 categorise (스타벅스 or 이디야(

```python

df_starbucks_ediya.loc[df_starbucks_ediya['상호명'].str.contains('스타벅스|starbucks|STARBUCKS'),'브랜드명']='스타벅스'
```

```python
df_starbucks_ediya.loc[df_starbucks_ediya['상호명'].str.contains('이디야|ediya|EDIYA'),'브랜드명']='이디야'
```

## folium 사용하여 지도상에 스타벅스 이디야 위치 나타내기

```python
import folium
```

```python
map = folium.Map(location=[df_starbucks_ediya['위도'].mean(), 
                           df_starbucks_ediya['경도'].mean()],
                           zoom_start=12, tiles='Stamen Toner')
for n in df_starbucks_ediya.index:
    # 팝업에 들어갈 텍스를 지정해 준다.
    popup_name = df_starbucks_ediya.loc[n,'상호명'] + ' - '\
    + df_starbucks_ediya.loc[n,'도로명']
    
    # 브랜드명에 따라 아이콘 색상을 다르게 준다.
    if df_starbucks_ediya.loc[n,'브랜드명']=='스타벅스':
        icon_color = 'green'
    else:
        icon_color = 'blue'
        
    folium.CircleMarker(
         location=[df_starbucks_ediya.loc[n,'위도'],df_starbucks_ediya.loc[n,'경도']],
         radius =3,
         popup = popup_name,
         color=icon_color,
         fill = True,
         fill_color=icon_color     
         
    ).add_to(map)
```

### 행정구역별 스타벅스와 이디야 매장 수 차이

```python
# reset_index () 기존 인덱스명을 초기화 
df_coffee_brand = pd.DataFrame(df_starbucks_ediya.groupby(['시군구명','브랜드명'])['상호명'].count()).reset_index()
```

```python
df_coffee_brand.head()
```

```python
df_coffee_brand.columns = ['구','브랜드명','매장수'] # count를 이용해 매장수 column 만들기
# 즉 상호명을 매장수로 치환 
```

```python
df_coffee_brand.head()
```

```python
plt.figure(figsize = (25,6),facecolor = 'white')
sns.barplot(data = df_coffee_brand, x = '구',y = '매장수', hue = '브랜드명')
```
