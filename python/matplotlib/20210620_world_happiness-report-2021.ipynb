```python
ladder_score_coun = df[['Country name','Ladder score','Standard error of ladder score']].sort_values('Ladder score', ascending = 'True',na_position='first')
```

```python
plt.figure(figsize=[20,10],facecolor = "white")
fig = plt.hist(ladder_score_coun['Ladder score'])
```

```python
plt.bar(ladder_score_coun['Ladder score'][0:10],ladder_score_coun['Country name'][0:10])
```

```python
plt.style.use('ggplot')
plt.figure(figsize = (10,5),facecolor = 'white')
fig = sns.barplot(data = ladder_score_coun[:10],x = 'Ladder score',y = 'Country name')
```
