# 폐암 수술 환자의 생존율 예측

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
import numpy as np
import tensorflow as tf
```

```python
# 실행할때마다 같은 결과를 출력하기 위해 설정
tf.random.set_seed(3)

# 데이터 셋 읽어오기
# 딥러닝에서는 넘파이 다차원 배열로 가져옴
data_set = np.loadtxt('data/ThoraricSurgery.csv',delimiter=',')

# 환자의 기록(target)과 수술 결과(label)를 x,y로 구분하여 저장
x = data_set[:,0:17]
y = data_set[:,17]

# (1) 딥러닝 모델 설정 : 딥러닝 구조 결정 (층을 결정, 입력값, 출력값 결정, 활성함수 적어주고 모델 설정 단계 )
model = Sequential()
model.add(Dense(30,input_dim=17, activation='relu')) # add가 층을 하나 추가 (30개가 출력 될것, 17개의 입력값이 들어갈것)
model.add(Dense(1, activation='sigmoid'))

# (2) 모델을 컴파일 (결과가 나왔을 때 어떤 손실함수로써 오차를 계산할건지, 어떤 옵티마이저를 써서 다시 한번 가중치를 수정할 것인지, 측정결과는 어떻게)
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy']) # 손실함수 이진구조

# (3) 모델 실행 (학습)
model.fit(x,y, epochs = 100, batch_size=10)
```

```python
print(np.__version__)
print(tf.__version__)
```

```python
print('\n Accuracy: %.4f' % (model.evaluate(x,y)[1]))
```

### (1) 딥러닝 모델 설정 : 딥러닝 구조 결정 (층을 결정, 입력값, 출력값 결정, 활성함수 적어주고 모델 설정 단계 )
model = Sequential()
model.add(Dense(30,input_dim=17, activation='relu')) # add가 층을 하나 추가 (30개가 출력 될것, 17개의 입력값이 들어갈것)
model.add(Dense(1, activation='sigmoid'))

* 딥러닝은 여러 층이 쌓여서 결과를 만들어 냄
* Sequential() / add () 를 사용해서 딥러닝 구조를 한층 한층 쌓아올림 
* add () : 필요한 층을 차례로 추가
* Dense() : 각 층이 어떤 특성을 가질지 옵션을 설정하는 역할 (조밀하게 모여 있는 집합)


### (2) 모델을 컴파일 (결과가 나왔을 때 어떤 손실함수로써 오차를 계산할건지, 어떤 옵티마이저를 써서 다시 한번 가중치를 수정할 것인지, 측정결과는 어떻게)
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy']) # 손실함수 이진구조

* compile 
    * 앞에서 지정한 모델이 효과적으로 구현될 수 있도록 여러가지 환경을 설정하면서 실행가능한 상태가 되도록 만들어주는 단계
    * 어떤 오차 함수를 사용할지 : loss
    * 어떤 최적화 방법을 사용할지 : optimizer
    * 수행 결과 평가는 어떤 방법을 사용할지 : metrics


### (3) 모델 실행 (학습)
model.fit(x,y, epochs = 100, batch_size=10)

* fit() : 데이터 입력하면 앞에서 설정한 대로 실행되고 결과 출력
* epochs : 학습 프로세스가 모든 샘플에 대해 한 번 실행하는 것
    * epochs = 100 : 각 샘플이 처음부터 끝까지 100번 재사용될때까지 반복한다는 의미
* batch_size : 전체 개수의  샘플을 n번으로 나누어서 처리하라는 의미
    * batch_size=10 : 전체 (e.g.150개)를 10번으로 나누면 샘플은 각 15개씩
    * batch size가 너무 크면 학습 속도가 느려지고 너무 작으면 실행 값의 편차가 생겨 전체 결과값이 불안정해 질 수 있음
    * 사용하는 메모리가 감당할 만큼의 배치 사이즈로 설정
    * 후에 프로젝트 할 때 사용해보기
